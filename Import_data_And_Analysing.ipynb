```python
import matplotlib.pyplot as plt
import matplotlib.style as style
from matplotlib import rcParams
import numpy as np
import pandas as pd

data1=pd.read_csv('ProjectpharmaTop10.csv')

plt.plot(data1.Company[0:5],data1.billions2018[0:5],'r.-',label=2018,marker="P")
plt.plot(data1.Company[0:5],data1.billions2017[0:5],'b.-',label=2017,marker="s")
plt.plot(data1.Company[0:5],data1.billions2016[0:5],'g.-',label=2016,marker="d")
plt.title('Pharma Company and their Revenue in last 3 years')
plt.xticks(rotation=20)
plt.ylabel('Revenue in USD billions')
plt.legend()
plt.show()
```
